## Text und Zeichenketten in Python

Bisher haben wir nur mit Zahlen gearbeitet, aber natürlich kann jede Programmiersprache auch mit Buchstaben und Text umgehen. Nicht-numerische Daten wie Text werden oft als Zeichenketten (Strings) bezeichnet; diese werden in Python zwischen zwei Anführungszeichen (" ") gesetzt.

```python
"Ich bin ein Text"
```

Auch Variablen können Zeichenketten zugewiesen werden. Ähnlich zu numerischen Variablen ist Python in der Lage, auch mit Zeichenketten-Variablen zu "rechnen".

```python
a = "Hallo"
b = "Welt"
print(a + b)
print(a + " " + b)
```

## Listen

Neben Zahlen und Zeichenketten können auch komplexere Daten in Form von **Listen** abgebildet werden. Prinzipiell kann man sich Listen auch als Splaten einer Tabelle vorstellen. Eine Liste kann also zum Beispiel eine Einkaufliste sein, auf der mehrere Dinge stehen, die du besorgen musst. Ein anderes Beispiel könnten die Noten aller Schüler:innen in der letzen Mathe-Klassenarbeit sein.

```python
einkaufsliste = ["Bananen", "Saft", "Chips"]

noten = [2, 1, 3, 2, 1, 1, 6, 4, 2, 2, 5, 1]
```

 In Python können Listen verbunden werden, aber man kann auch innerhalb mit einer Liste Rechnen:

```python
#Listen zusammenfügen
einkaufliste2 = ["Brötchen", "Milch"]
print(einkaufsliste + einkaufliste2)

#Länge einer Liste ermitteln
anzahl = len(noten)

#Summe aller Elemente einer Liste
summe = sum(noten)

#Noten-Durchschnitt
print("Anzahl Klassenarbeiten: ", anzahl)
print("Summe aller Noten: ", summe)
print("Noten-Durchschnitt: ", summe / anzahl)
```

Nach den Grundlagen wollen wir endlich mal ein richtiges Programm starten.

```python
from tkinter import *  
from tkinter import messagebox 

buttons       = [[None] * 3, [None] * 3, [None] * 3]
felder_fertig = 0

def clicked_spielfeld(r, c):
    global felder_fertig
    global buttons

    # bereits geklickt?
    if buttons[r][c].cget("text") != "":
        return

    # setze Zeichen
    if felder_fertig % 2 == 0:
        buttons[r][c].configure(text = "X")
    else:
        buttons[r][c].configure(text = "O")
    felder_fertig = felder_fertig + 1

    # kontrolliere zum Schluss, ob das Rätsel gewonnen oder unentschieden ist
    gewonnen = checke_gewonnen(r, c)
    if (not gewonnen) and felder_fertig == 9:
        messagebox.showinfo("Unentschieden", "Unentschieden!")

def checke_gewonnen(zeile, spalte):
    global buttons
    spieler = buttons[zeile][spalte].cget("text")

    if buttons[zeile][0].cget("text") == buttons[zeile][1].cget("text") == buttons[zeile][2].cget("text"):
        messagebox.showinfo("Winner", spieler + " Won")
        return True

    if buttons[0][spalte].cget("text") == buttons[1][spalte].cget("text") == buttons[2][spalte].cget("text"):
        messagebox.showinfo("Winner", spieler + " Won!")
        return True

    if (zeile, spalte) in [(0, 0), (1, 1), (2, 2)]:
        if buttons[0][0].cget("text") == buttons[1][1].cget("text") == buttons[2][2].cget("text"):
            messagebox.showinfo("Winner", spieler + " Won!")
            return True

    if (zeile, spalte) in [(0, 2), (1, 1), (2, 0)]:
        if buttons[0][2].cget("text") == buttons[1][1].cget("text") == buttons[2][0].cget("text"):
            messagebox.showinfo("Winner" , spieler + " Won!")
            return True

    return False

def clicked_neu():
    global felder_fertig
    global buttons

    felder_fertig = 0
    for z in range(3):
        for s in range(3):
            buttons[z][s].configure(text = "")


# Grafische Benutzeroberfläche erstellen...
root = Tk()
root.title("Tic Tac Toe")
root.resizable(0,0)

# erstelle die 9 Buttons des Spielfeldes
for i in range(3):
    for j in range(3):
        buttons[i][j] = Button(
            height = 4,
            width = 8,
            font = ("Helvetica", "20"),
            command = lambda r=i, c=j: clicked_spielfeld(r, c)
        )
        buttons[i][j].grid(row = i, column = j)
# erstelle "Neues Spiel"-Button
btn_neu = Button(text = "Neues Spiel", command = clicked_neu)
btn_neu.grid(row = 1, column = 3)

mainloop()
```
