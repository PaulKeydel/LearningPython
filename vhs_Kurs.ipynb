## Schleifen und bedingte Anweisungen

Ein weiteres zentrales Werkzeug in Programmiersprachen sind die **Schleifen**. Eine Schleife ermöglicht es uns, bestimmte Operationen beliebig oft auszuführen. Es gibt verschiedene Arten von Schleifen, wir beschränken uns hier auf die **for-Schleifen**. Diese sind dafür gemacht, durch eine Liste durchzugehen - Element für Element. Am besten verdeutlichen wir uns das mit unserer Noten-Liste:

```python
noten = [2, 1, 3, 2, 1, 1, 6, 4, 2, 2, 5, 1]
for note in noten:
    print(note)

#Summe berechnen
summe = 0
for note in noten:
    summe = summe + note
print("Summe mit Schleife: ", summe)
print("Summe mit sum-Funktion: ", sum(noten))
```

Es gibt noch eine zweite Möglichkeit mit einer for-Schleife eine Liste durchzugehen, indem die Schleife nicht direkt über die Listenelemente läuft, sondern über die Positionen:

```python
testliste = ["element1", "element2", "element3"]

for position in [0, 1, 2]:
    print(testliste[position])

for position in range(3):
    print(testliste[position])

for position in range(len(testliste)):
    print(testliste[position])
```

Oft hat man die Situation, dass man etwas nur machen will, wenn eine bestimmte Bedingung eingetreten ist. Das kann z.B. passieren, wenn uns interessiert, wie viele Klassenarbeiten es mit der Note 2 gibt. Für so eine Aufgabe brauchen wir **bedingte Anweisungen**, die in Python mit *if* beginnen (zu deutsch "wenn"). Wenn also eine bestimmte Bedingung erfüllt ist, wird nur in diesem Fall Code ausgeführt.
Beachte: Wir vergleichen nicht mit "=", sondern mit "==" und "!=" (ungleich).

```python
eingabe = ""
if eingabe == "":
    print("Variable ist leer")
```

Uns interessiert jetzt, wie viele Klassenarbeiten es mit der Note 2 gibt:

```python
gute_arbeiten = 0
for note in noten:
    if note == 2:
        print("Das ist eine gute Arbeit")
        gute_arbeiten = gute_arbeiten + 1
    if note != 2:
        print("Die Arbeit ist besser oder schlechter als 2")

print("Anzahl der Arbeiten mit der Note 2: ", gute_arbeiten)
# eine Abkürzung mit dem count-Befehl
print("Probe: ", noten.count(2))
```

Nach den ganzen Grundlagen kommen wir endlich zu unserem TicTacToe. Bevor wir es allerdings auf dem Computer spielen können, müssen wir erstmal überlegen, wie wir das Spiel im Computer nachbilden. Beginnen wir mit der Frage, wie wir ein 3x3 Spielfeld speichern könnnen? Das geht relativ leicht mit einer geschachetelten Liste. Jede Zeile des Spielfeldes ist einer 3er-Liste und die drei Zeilen befinden sich in einer übergeordenten Liste. Das obige TicTacToe ließe sich so in Python übersetzen:

```python
quiz = [
    ["O", "O", "O"],
    ["", "O", "X"],
    ["", "X", "X"]
]
```

Mit dem Computer müssen wir es schaffen, drei Fragen zu beantworten:
1. Wer ist als nächstes dran? (wenn nicht gewonnen und nicht unentschieden)
2. Ist das Spiel gewonnen?
3. Ist das Spiel unentschieden?

## Nächster Spieler?

```python
quiz = [
    ["", "X", "X"],
    ["O", "", "X"],
    ["", "", "O"]
]

def naechster_spieler(quiz):
    anzahl_x = quiz[0].count("X") + quiz[1].count("X") + quiz[2].count("X") #sum([sum([k == "X" for k in row]) for row in quiz])
    anzahl_o = quiz[0].count("O") + quiz[1].count("O") + quiz[2].count("O") #sum([sum([k == "O" for k in row]) for row in quiz])
    if anzahl_x == anzahl_o:
        return "X"
    else:
        return "O"

print("Nächste*r Spieler*in: ", naechster_spieler(quiz))
```

## Spiel gewonnen?

```python
quiz = [
    ["", "", "O"],
    ["X", "X", "X"],
    ["O", "", ""]
]

def ist_gewonnen(quiz):
    for zeile in range(3):
        if quiz[zeile][0] == "":
            continue
        if quiz[zeile][0] == quiz[zeile][1] == quiz[zeile][2]:
            return True

    for spalte in range(3):
        if quiz[0][spalte] == "":
            continue
        if quiz[0][spalte] == quiz[1][spalte] == quiz[2][spalte]:
            return True

    if quiz[1][1] != "":
        if quiz[0][0] == quiz[1][1] == quiz[2][2]:
            return True

        if quiz[0][2] == quiz[1][1] == quiz[2][0]:
            return True

    return False

print("Gewonnen? ", ist_gewonnen(quiz))
```

## Spiel unentschieden?

```python
quiz = [
    ["X", "O", "X"],
    ["X", "O", "O"],
    ["O", "X", "X"]
]

def ist_ausgefuellt(quiz):
    felder_fertig = sum([sum([k != "" for k in row]) for row in quiz])
    if felder_fertig == 9:
        return True
    else:
        return False

print("Gewonnen? ", ist_gewonnen(quiz))
print("Unentschieden? ", not(ist_gewonnen(quiz)) and ist_ausgefuellt(quiz))
```

# 3.Modul: Endspurt TicTacToe - Benutzeroberflächen in und mit Python

Nach den Grundlagen wollen wir endlich mal etwas Grafik ins Spiel bringen.

```python
import turtle

t = turtle.Pen()

t.forward(20)
```

```python
from tkinter import *  
from tkinter import messagebox 

buttons = [[None] * 3, [None] * 3, [None] * 3]

def clicked_spielfeld(r, c):
    global buttons

    # bereits geklickt?
    if buttons[r][c].cget("text") != "":
        return

    spielfeld = [[""] * 3, [""] * 3, [""] * 3]
    for z in range(3):
        for s in range(3):
            if buttons[z][s].cget("text") == "":
                continue
            spielfeld[z][s] = buttons[z][s].cget("text")

    # bereits gewonnen?
    if ist_gewonnen(spielfeld):
        return

    spieler_zug = naechster_spieler(spielfeld)

    buttons[r][c].configure(text = spieler_zug)
    spielfeld[r][c] = spieler_zug

    # kontrolliere zum Schluss, ob das Rätsel gewonnen oder unentschieden ist
    gewonnen = ist_gewonnen(spielfeld)
    if gewonnen:
        messagebox.showinfo("Winner", spieler_zug + " hat gewonnen!")
    elif ist_ausgefuellt(spielfeld):
        messagebox.showinfo("Unentschieden", "Unentschieden!")

def clicked_neu():
    global buttons
    for z in range(3):
        for s in range(3):
            buttons[z][s].configure(text = "")


# Grafische Benutzeroberfläche erstellen...
root = Tk()
root.title("Tic Tac Toe")
root.resizable(0,0)

# erstelle die 9 Buttons des Spielfeldes
for i in range(3):
    for j in range(3):
        buttons[i][j] = Button(
            height = 4,
            width = 8,
            font = ("Helvetica", "20"),
            command = lambda r=i, c=j: clicked_spielfeld(r, c)
        )
        buttons[i][j].grid(row = i, column = j)
# erstelle "Neues Spiel"-Button
btn_neu = Button(text = "Neues Spiel", command = clicked_neu)
btn_neu.grid(row = 1, column = 3)

mainloop()
```
